```python
Skewness  = pd.DataFrame({'Skewness':[stats.skew(df.pclass),stats.skew(df.survived),stats.skew(df.sibsp),stats.skew(df.parch)]},index=['pclass','survived','sibsp','parch'])
Skewness  # measure the skewness of the required columns
```

#### highest age in this ship

```python

top_ages = df['age'].value_counts().head(5)

# Create a bar plot using Seaborn
sns.barplot(x=top_ages.index, y=top_ages.values)
plt.title('Top 5 Most Common Ages')
plt.xlabel('Age')
plt.ylabel('Count')

# Display the plot
plt.show()
```

- Above graphs shows that top 5 age which is present in titanic ship between 18-30, in which 24 age highest in the ship.

### Sex ratio in the ship

```python
top_sex = df['sex'].value_counts()

# Create a bar plot using Seaborn
sns.barplot(x=top_sex.index, y=top_sex.values)
plt.title(' Sex ratio')
plt.xlabel('Sex')
plt.ylabel('Count')

# Display the plot
plt.show()
```

- Highest sex ratio present in the ship is male

### who survived most Gender wise

```python
# Calculate survival rates by gender
survival_by_gender = df.groupby('sex')['survived'].mean().reset_index()

# Create a bar plot to visualize survival rates by gender
plt.figure(figsize=(10, 4))
sns.barplot(x='sex', y='survived', data=survival_by_gender)
plt.title('Survival Rates by Gender')
plt.xlabel('Gender')
plt.ylabel('Survival Rate')
plt.ylim(0, 1)  # Set y-axis limit to be between 0 and 1 (for percentages)
plt.show()
```

#### above graph we said that female ratio is high as compared to man who survived

### which age is survived most

```python
# Calculate survival rates by gender
survival_by_gender = df.groupby('age')['survived'].mean().reset_index().head()

# Create a bar plot to visualize survival rates by gender
plt.figure(figsize=(15, 10))
sns.barplot(x='age', y='survived', data=survival_by_age)
plt.title('Survival Rates by age')
plt.xlabel('age')
plt.ylabel('Survival Rate')

plt.show()
```

### correlation

```python
correlation_matrix = df.corr()
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm')
plt.title('Correlation Heatmap')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.violinplot(data=df, x='pclass', y='age', hue='survived', split=True)
plt.title('Age Distribution by Passenger Class and Survival')
plt.xlabel('Passenger Class')
plt.ylabel('Age')
plt.show()
```
